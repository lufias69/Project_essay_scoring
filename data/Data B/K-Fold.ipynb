```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from joblib import dump
import numpy as np
from sklearn import svm
import json
```

```python
data = pd.read_excel('data/Data_B_Bersih_2.xlsx')
label = data['new_label']
res_id = data['res-id']
# y = [str(x) for x in label]
respon = data['respon'].tolist()
# respon
y      = np.array(label)
res_id = np.array(res_id)
data.head()
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(respon)
```

```python
K = 10
it= 0
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
kf= KFold(n_splits=K) 
cSVM = svm.SVC(kernel='rbf', gamma=0.7, C=2.6)
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train = y[train_index]
    y_test  = y[test_index]
    model = cSVM.fit(X_train, y_train)
    P_SVM = cSVM.predict(X_test)
    res_id_test = res_id[test_index]
    P_SVM = list(P_SVM)
    dict_={
        "RES_ID":res_id_test,
        "LABEL":P_SVM
    }
    it+=1
    print('   Iterasi Ke-'+str(it))
    akurasi = accuracy_score(y_test, P_SVM)
    print("\tAkurasi :",akurasi)
    print("")
    df = pd.DataFrame.from_dict(dict_)
    df.to_excel("hasil/train/json/predictions_dev_"+str(it)+"_.xlsx")
    
    list_hasil = list()
    for r, p in zip(res_id_test, P_SVM):
        dic={
            'RES_ID':r,
            'LABEL':int(p)
        }
        list_hasil.append(dic)
    with open('hasil/train/exce/predictions_dev_'+str(it)+'_.json', 'w') as outfile:
        json.dump(list_hasil, outfile)
```
