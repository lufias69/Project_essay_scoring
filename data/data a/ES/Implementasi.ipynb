```python
def trufalse(list1, list2):
    hasil = list()
    for i, j in zip(list1, list2):
        if i == j:
            hasil.append("True")
        else:
            hasil.append("False")
    return hasil
```

# Implementasi

```python
f = True #fitur hanya berasal dari 
c = False

cosine_skor = list()
cosine_qualifikasi = list()
cosine_similarity = list()
cosine_status = list()
cosine_cwm = list()

dice_skor = list()
dice_qualifikasi = list()
dice_similarity = list()
dice_status = list()
dice_cwm = list()

jaccard_skor = list()
jaccard_qualifikasi = list()
jaccard_similarity = list()
jaccard_status = list()
jaccard_cwm = list()

for ix, j in enumerate(jawaban):
    hasil = sc.scoring_cosine(j,kuncijawaban_list,skor_list, fitur=f, char = c)
    cosine_skor.append(hasil[0])
#     cosine_qualifikasi.append(hasil[1])
    cosine_similarity.append(hasil[1])
    cosine_status.append(hasil[2])
    cosine_cwm.append(sc.cek_mised(hasil[2]))
    
    hasil = sc.scoring_dice(j,kuncijawaban_list,skor_list, fitur=f, char = c)
    dice_skor.append(hasil[0])
#     dice_qualifikasi.append(hasil[1])
    dice_similarity.append(hasil[1])
    dice_status.append(hasil[2])
    dice_cwm.append(sc.cek_mised(hasil[2]))
    
    hasil = sc.scoring_jaccard(j,kuncijawaban_list,skor_list, fitur=f, char = c)
    jaccard_skor.append(hasil[0])
#     jaccard_qualifikasi.append(hasil[1])
    jaccard_similarity.append(hasil[1])
    jaccard_status.append(hasil[2])
    jaccard_cwm.append(sc.cek_mised(hasil[2]))

    if ix %10==0 and ix != 0:
        if ix !=0:
            print(ix, end=" ")
    else:
        print(".", end="")
```

# Rouge

```python

# data_rouge = ["cosine_cwm","dice_cwm","jaccard_cwm"]
data_rouge = ["cosine_cwm"]#,"dice_cwm","jaccard_cwm"]
nama       = ["Cosine"]#,"Dice","Jaccard"]

metode = list()
precision = list()
recall = list()
f1_score = list()

correctl = list()
missedl = list()
wrongl = list()

for i, j in zip(data_rouge, nama):
    dataf = dataframe_hasil[i].tolist()
    correct = 0
    missed  = 0
    wrong   = 0
    for km in dataf:
        if km == "correct":
            correct +=1
        elif km == "missed":
            missed +=1
        elif km =="wrong":
            wrong +=1
        else:
            print(km)
    
    pres = correct/(correct+wrong)
    precision.append(pres)
    rec =  correct/(correct+missed)
    recall.append(rec)
    fm = 2*pres*(rec/(rec+pres))
    f1_score.append(fm)
    
    correctl.append(correct)
    missedl.append(missed)
    wrongl.append(wrong)
    
dict1 = {
        "metode":nama,
        "correct":correctl,
        "missed":missedl,
        "wrong":wrongl,
        }

dict2 = {
        "metode":nama,
        "precision":precision,
        "recall":recall,
        "f-Measure":f1_score,
        }
df_cwm = pd.DataFrame.from_dict(dict1).set_index('metode')
df_f_score = pd.DataFrame.from_dict(dict2).set_index('metode')
```
