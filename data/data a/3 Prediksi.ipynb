```python
import pandas as pd
from model import prediksi
import json
```

# Load Data Uji

```python
data = pd.read_excel("data/data_dev_A_labeled_stemer(True).xlsx")
# print(data)
res_id = data['res_id']
respon = data['respon_old']
label = data['new_label']
# data
```

# Proses Prediksi

```python
# prediksi.prediksi_single("Beradaptasi dengan lingkungan")
```

```python
hasil = prediksi.prediksi_list(respon)
```

# Simpan hasil prediksi

```python
prediksi_ = hasil['prediksi'].values.astype('U')
list_hasil = list()
for r, p in zip(res_id, prediksi_):
    dic={
        'RES_ID':r,
        'LABEL':p
    }
    list_hasil.append(dic)
json.dumps(list_hasil)

with open('hasil/predictions_dev.json', 'w') as outfile:
    json.dump(list_hasil, outfile)
```

```python
prediksi = hasil['prediksi'].tolist()
dict_ = {
    'RES_ID':res_id,
    'LABEL':prediksi,
    'RESPONSE':respon,
}
df=pd.DataFrame.from_dict(dict_)
df.to_excel("hasil/Hasil_all.xlsx")
df
```

```python
from sklearn.metrics import classification_report
prediksi_ = [str(x) for x in prediksi_]
label = [str(x) for x in label]
y_true = label
y_pred = prediksi_
target_names = ['0','1']
prediksi_
print(classification_report(y_true, y_pred, target_names=target_names))
```
